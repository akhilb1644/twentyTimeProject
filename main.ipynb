```python
import plotly.graph_objects as pgo
import pandas as pd
import json

class Attraction:
    def __init__(self,name,interest,state,lat,long):
        self.name = name
        self.interest = interest
        self.state = state
        self.lat = lat
        self.long = long
    
    def __str__(self):
        return f"Name: {self.name} Interest {self.interest} State {self.state} Latitude {self.lat} Longitude {self.long}"
        
lats = list()
lons = list()
texts = list()

lats1 = list()
lons1 = list()
texts1 = list()

lats2 = list()
lons2 = list()
texts2 = list()
        
mapAttractions = list()
states = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
```

```python
with open('hikingCoords.json') as c:
    coords = json.load(c)
    
coordsList = list()
for coord in coords.values():
    coordsList.append(coord)
    
print(coordsList)
print(len(coordsList))
```

```python
with open('hiking.json') as h:
    attractions = json.load(h)
    
attractionList = list()
for attraction in attractions.values():
    attractionList.append(attraction)
    
print(attractionList)
print(len(attractionList))
```

```python
sIndex = 0
for state in attractionList:
    index = 0
    for attraction in state:
        a = Attraction(attraction,"Hiking",states[sIndex],coordsList[sIndex][index*2],coordsList[sIndex][index*2+1])
        mapAttractions.append(a)
        index += 1
    sIndex += 1
    
```

```python
visits = list()

with open('khrushchev.json') as k:
    cold = json.load(k)
    
with open('khrushchevCoords.json') as kc:
    coldCoords = json.load(kc)
    
kCoords = list(coldCoords.values())
kStates = list(cold.keys())
index = 0
for visit in cold.values():
    v = 0
    for value in visit:
        visits.append(Attraction(value,"Khrushchev's Visit",kStates[index],kCoords[index][v*2],kCoords[index][v*2+1]))
        v += 1
    index += 1
    
print(visits)
```

```python
karts = list()

with open('goKarts.json') as oka:
    rub = json.load(oka)
    
with open('goKartsCoords.json') as rts:
    rubCoords = json.load(rts)
    
gCoords = list(rubCoords.values())
gStates = list(rub.keys())
f = 0
for goKart in rub.values():
    w = 0
    for eman in goKart:
        karts.append(Attraction(eman,"GoKarts",gStates[f],gCoords[f][w*2],gCoords[f][w*2+1]))
        print(gStates[f])
        w += 1
    f += 1
    
for kart in karts:
    print(kart.name)
```

```python
for at in mapAttractions:
    lats.append(at.lat)
    lons.append(at.long)
    texts.append(at.name)
    
for att in visits:
    lats1.append(att.lat)
    lons1.append(att.long)
    texts1.append(att.name)
    
for attr in karts:
    lats2.append(attr.lat)
    lons2.append(attr.long)
    texts2.append(attr.name)

mapdata = [
pgo.Scattergeo(
    lat=lats,
    lon=lons,
    mode='markers',
    text = texts,
    name = "Hiking Attractions"
),pgo.Scattergeo(
    lat=lats1,
    lon=lons1,
    mode='markers',
    text = texts1,
    name = "Khrushchev's visit(simpified)"
),pgo.Scattergeo(
    lat=lats2,
    lon=lons2,
    mode='markers',
    text = texts2,
    name = "GoKart Locations"
)]

fig = pgo.Figure(data = mapdata)

fig.update_geos(center_lat=44.96504742388492,center_lon=-103.77264795846477,projection_scale=2.5,showcountries=True,showsubunits=True,showrivers=True)

fig.show()
fig.write_html('map.html')
```
